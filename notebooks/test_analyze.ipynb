```python
# get environment variables from .env file
import os
import requests
from environment_util import load_env_file
load_env_file("/Users/scooper/code/stock_advisor/.env")

BEARER_TOKEN = os.environ.get("BEARER_TOKEN")
```

```python
import requests

# URL of the API endpoint
url = 'http://localhost:3333/analyze/'

# Parameters for the API call
params = {
    'symbol': 'AAPL'
}

# Making a GET request to the API
response = requests.get(url, params=params)

# Checking if the request was successful
if response.status_code == 200:
    # Printing the response content
    print(response.json()["results"])
else:
    # Printing an error message if the request was not successful
    print(f"Failed to fetch data. Status code: {response.status_code}")
```

```python
import requests

# URL of the API endpoint (production)
url = 'https://data.aitickerchat.com/analyze/'

# Headers to include the bearer token
headers = {
    'Authorization': f'Bearer {BEARER_TOKEN}'
}

# Parameters for the API call
params = {
    'symbol': 'AAPL'
}

# Making a GET request to the API with headers
response = requests.get(url, headers=headers, params=params)

# Checking if the request was successful
if response.status_code == 200:
    # Printing the response content
    print(response.json()["results"])
else:
    # Printing an error message if the request was not successful
    print(f"Failed to fetch data. Status code: {response.status_code}")
```
